```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys
```

```python
def get_signal(file_name):
    micro = wave.open(file_name, "r")

    signal = micro.readframes(-1)
    signal = np.fromstring(signal, "Int16")
    
    micro.close()
    
    return signal
```

```python
signal0 = get_signal("convolutionAttention0.wav")
signal1 = get_signal("convolutionAttention1.wav")
signal2 = get_signal("convolutionAttention2.wav")

plt.figure(figsize=(20,10))
plt.title('Signal on microphone 1, 2, 3 in big room')
plt.plot(signal0, color='r')
plt.plot(signal1, color='g')
plt.plot(signal2, color='b')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
def get_values(file_name):
    f = open(file_name, "r")
    vals =[]
    for val in f:
        vals.append(float(val))
        
    return vals
```

```python
x = get_values("impulseResponseBigRoom8192/xImpulseResponse.txt")
y = get_values("impulseResponseBigRoom8192/yImpulseResponse.txt")

plt.figure(figsize=(20,10))
plt.title('Impulse response third microphone in big room')
plt.plot(x, y, color='r')
plt.xlabel("Time")
plt.ylabel("Pressure")
plt.show()
```
