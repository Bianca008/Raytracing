```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys
import math
import pandas as pd 
```

```python
def get_signal(file_name):
    micro = wave.open(file_name, "r")

    signal = micro.readframes(-1)
    signal = np.fromstring(signal, "Int16")
    
    micro.close()
    
    return signal
```

```python
def plot_sound():
    signal0 = get_signal("convolutionAttention0.wav")
    signal1 = get_signal("convolutionAttention1.wav")
    signal2 = get_signal("convolutionAttention2.wav")

    plt.figure(figsize=(20,10))
    plt.title('Semnalul pe microfoanele 0, 1, 2 în Camera 2')
    plt_signal_1, = plt.plot(signal0, color='r', label='Microfon 0')
    plt_signal_2, = plt.plot(signal1, color='g', label='Microfon 1')
    plt_signal_3, = plt.plot(signal2, color='b', label='Microfon 2')
    plt.legend(loc='upper right')
    #second_legend = plt.legend(handles=[plt_signal_2])
    #third_legend = plt.legend(handles=[plt_signal_3])
    plt.xlabel("Timp(1e+3s)")
    plt.ylabel("Amplitudine(Pa)")
    plt.show()
```

```python
def get_values(file_name):
    f = open(file_name, "r")
    vals =[]
    for val in f:
        vals.append(float(val))
        
    return vals
```

```python
def read_values(file_name, headers):
    df = pd.read_csv(file_name) 

    x = df[headers[0]]
    y = df[headers[1]]
    
    return x, y
```

```python
def plot_impulse_response():
    x = get_values("impulseResponseBigRoomMicro2-16384/xImpulseResponse.txt")
    y = get_values("impulseResponseBigRoomMicro2-16384/yImpulseResponse.txt")
    y = [el/(2*16384) for el in y]
    x_ir, y_ir = read_values("impulseResponseBigRoomMicro2-16384/ir_departe_long.csv", ['Time(sec)', 'Pressure(Pa)'])
    x_ir = [el/2 for el in x_ir]
    
    plt.figure(figsize=(20,10))
    plt.title('Funcția de răspuns la impuls în Camera 2 - Microfon 2')
    print(len(x), len(x_ir))
    #plt.plot(x_ir, y_ir, color='b', label="Simcenter")
    plt.plot(x, y, color='r', label="Model acustic")
    
    plt.legend(loc='upper right')
    plt.xlabel("Timp(s)")
    plt.ylabel("Presiune(Pa)")
    plt.show()
```

```python
def plot_frequency_magnitude():
    x = get_values("frResponseBigRoomMicro2-8192/xFrequencyMagnitude.txt")
    y = get_values("frResponseBigRoomMicro2-8192/yFrequencyMagnitude.txt")
    x_fm, y_fm = read_values("frResponseBigRoomMicro2-8192/fr_magn_departe_long.csv", ['Frequency(Hz)','Pressure(Pa)'])
    #y_fm = [el + 30 for el in y_fm]
    
    print(len(x), len(x_fm))
    plt.figure(figsize=(20,10))
    plt.title('Funcția de răspuns în frecvență în Camera 2 - Frecvență-Magnitudine - Microfon 2')
    plt.plot(x_fm, y_fm, color='b', label='Simcenter')
    plt.plot(x, y, color='r', label='Model acustic')  
    
    plt.legend(loc='upper right')
    plt.xlabel("Frecvență(Hz)")
    plt.ylabel("Presiune(Pa)")
    plt.show()
```

```python
def plot_frequency_phase():
    x = get_values("frResponseBigRoomMicro2-8192/xFrequencyPhase.txt")
    y = get_values("frResponseBigRoomMicro2-8192/yFrequencyPhase.txt")
    #x = [el/10000 for el in x]
    x_fp, y_fp = read_values("frResponseBigRoomMicro2-8192/fr_phase_departe_long.csv", ['Frequency(Hz)', 'Angle(degrees)'])
    y_fp +=180
    y_fp *= math.pi/180.0
    #y_fp += 9.5

    plt.figure(figsize=(20,10))
    plt.title('Funcția de răspuns în frecvență în Camera 2 - Frecvență-Fază - Microfon 2')
    plt.plot(x_fp, y_fp, color='b', label='Simcenter')
    plt.plot(x, y, color='r', label='Model acustic')
   
    plt.legend(loc='upper right')
    plt.xlabel("Frecvență(Hz)")
    plt.ylabel("Fază(radiani)")
    plt.show()
```

```python
def plot_time_echogram():
    x = get_values("timeEchogramBigRoomMicro2-8192/xTimeMagnitude.txt")
    y = get_values("timeEchogramBigRoomMicro2-8192/yTimeMagnitude.txt")
    x_tm, y_tm = read_values("timeEchogramBigRoomMicro2-8192/eco_departe_long.csv", ['Time(sec)', 'Pressure(Pa)'])

#     print(x[0],xIr[0],y[0],yIr[0])
    print(len(x), len(x_tm))
    fig = plt.figure(figsize=(20,10))
    ax = fig.add_axes([0,0,1,1])
    plt.title('Ecogramă timp 0 Hz în Camera 2 - Microfon 2')
    ax.bar(x_tm, y_tm, color='b',width=0.002, label='Simcenter')
    ax.bar(x,y,color='r',width=0.002, label='Model acustic')
        
    plt.legend(loc='upper right')
    plt.xlabel("Timp(s)")
    plt.ylabel("Presiune(Pa)")
    plt.show()
```

```python
def run():
    plot_time_echogram()
    plot_frequency_magnitude()
    plot_frequency_phase()
    plot_impulse_response()
    plot_sound()
```
